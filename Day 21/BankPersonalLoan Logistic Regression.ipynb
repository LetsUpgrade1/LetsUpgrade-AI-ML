```python
Y = data.PersonalLoan
Y.head()
```

```python
X = data[['ID', 'Age', 'Experience', 'Income', 'ZIPCode', 'Family', 'CCAvg',
       'Education', 'Mortgage', 'SecuritiesAccount', 'CDAccount', 'Online', 'CreditCard']]
X.head()
```

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
```

```python
data["CCAvg"]=le.fit_transform(data["CCAvg"])
data.head()
```

```python
import statsmodels.api as sm
```

```python
X1 = sm.add_constant(X)
```

```python
logistic = sm.Logit(Y,X1)
```

```python
result = logistic.fit()
```

```python
result.summary()
```
